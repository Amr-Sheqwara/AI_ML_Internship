# Day 1 — Environment Setup & the Jupyter Workflow

**Goal:** set up a reproducible Python data-science environment (venv + pip), get comfortable with the Jupyter code/markdown cell workflow, and commit the result to Git — the same environment reused for the rest of the 400-hour program.

---
## Step 1: Create and activate a virtual environment

A virtual environment keeps this project's package versions isolated from every other project on the machine, so dependencies never conflict.

```bash
python3 -m venv .venv
source .venv/bin/activate   # Linux/macOS
.venv\Scripts\activate     # Windows
```

---
## Step 2: Install NumPy, Pandas, and Matplotlib.
Now verifying the installation by checking their versions.

````python
import numpy as np
import pandas as ps 
import matplotlib as plt 

print("NumPy version:", np.__version__)
print("Pandas version:", ps.__version__)
print("Matplotlib version:", plt.__version__)
````

---
## Step 4: Freeze the environment to `requirements.txt`

Recording the exact installed versions means this environment can be recreated identically on any machine later in the program.

```bash
pip freeze > requirements.txt
```

The resulting `requirements.txt` is committed alongside this notebook in this folder.
