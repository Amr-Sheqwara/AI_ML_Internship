### Step 1: Statistical Function

**what it does**: This function iterates through a list of numbers to manually calculate the minimum, maximum, and average without using Python's built-in math functions, returning the result as a structured dictionary.

**why do we use it**: Functions package reusable logic. A well-named function with clear parameters and a single responsibility is the
foundation of clean, testable data pipelines.

```python
def NumberDictionary(values):
    current_min = values[0]
    current_max = values[0]
    count = 0
    sum = 0
    for value in values:
        count += 1
        sum = sum + value
        
        if value < current_min:
            current_min = value
        elif value > current_max:
            current_max = value
        
    
    return {"max": current_max, "min": current_min, "mean": sum / count}
```

```python
x = [1, 2, 3, 4 ,5]
NumberDictionary(x)
```

## Step 2: Even number filter

 **what it does**: A for-loop that filters even numbers as a single list of comprehension 

 **why do we use it**: A list comprehension builds a new list in one readable line, replacing a multi-line loop. This is idiomatic ("Pythonic")
code and appears constantly in real data-science work

```python
def EvenOdd(values):
    result = [value for value in values if value % 2 == 0]
    return result
```

```python
x = [1, 2, 3, 4 ,5]
EvenOdd(x)
```

## Step 3: DataRecord Class Initialization

**what it does**: A small class representing a data record with at least two attributes and one method

**why do we use it**: using a class allows us to bundle data (attributes) and behavior (methods) together.

```python
class DataPoint:
   def __init__(self, x, y):
       self.x = x
       self.y = y
   def distance_from_origin(self):
       return (self.x**2 + self.y**2) ** 0.5

p = DataPoint(2, 4)
print(p.distance_from_origin())
```
