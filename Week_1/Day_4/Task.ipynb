## Step 1: Loading the dataset
This loads the CSV dataset into a DataFrame and reports its shape, columns, and dtypes.

```python
import pandas as pd
df = pd.read_csv("tested.csv")
print("Dataset Shape:", df.shape) 
print("----------------------")
print("Dataset Info:")
df.info() 
```

---
## Step 2: Identify Missing Values
Count missing values per column and handle them (fill or drop) with a documented justification

```python
df.isnull().sum()
```

Here we dropped the column Cabin because it had too many missing values (327 missing values filling it would be unoptimized).                  
We filled the whole column Age with the mean of the column (has only 86 missing values it would be best to fill it).                    
we dropped the missing values in the column Fare because it was only 1 missing value. 

```python
df = df.drop(columns=["Cabin"])
df["Age"] = df["Age"].fillna(df["Age"].mean())
df = df.dropna(subset=["Fare"])
```

```python
df.isnull().sum()
```

---
## Step 3 + 4: Filter the data to a meaningful subset & Use groupby
Filtered the data to a meaningful subset and described what i found then Used groupby to compute an aggregate statistic per category and interpret the result in a Markdown cell.

```python
adults_df = df[df["Age"] > 18]
print(f"We filtered down to {len(adults_df)} adult passengers.")

adult_count = adults_df.groupby("Sex")["Age"].count()
print("\nAdult Count by Sex:\n", adult_count)
```
